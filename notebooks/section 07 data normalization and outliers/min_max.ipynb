```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## Min-Max Scaling
Transforming data to a new range scale
Unity normed scale is min-max of 0 to 1

Formula: x(i) = (x(i) - min(x)) / (max(x) - min(x))

```python
data = np.random.randn(1000)

min = np.min(data)
max = np.max(data)
mean = np.mean(data)

fig_a, ax_a = plt.subplot_mosaic([['plot', 'hist']], figsize=[12, 4])

ax_a['plot'].plot(data, '.')
ax_a['plot'].set_ylabel('Data value')
ax_a['plot'].set_xlabel('Data index')

ax_a['hist'].hist(data, bins = 30)
ax_a['hist'].set_ylabel('Count')
ax_a['hist'].set_xlabel('Data value')

fig_a.suptitle(f"Random Normal Sample\nMin = {np.round(min, 2)}, Max = {np.round(max, 2)}")
fig_a.show()

```

```python
unit_norm = (data - min) / (max - min)
un_min = np.min(unit_norm)
un_max = np.max(unit_norm)
un_mean = np.mean(unit_norm)

fig_b, ax_b = plt.subplot_mosaic([['plot', 'hist']], figsize = (12, 4))

ax_b['plot'].plot(unit_norm, '.')
ax_b['plot'].set_ylabel('Data value')
ax_b['plot'].set_xlabel('Data index')

ax_b['hist'].hist(unit_norm, bins = 30)
ax_b['hist'].set_ylabel('Count')
ax_b['hist'].set_xlabel('Data value')

fig_b.suptitle(f"Unity Normal Sample\nMin = {np.round(un_min, 2)}, Max = {np.round(un_max, 2)}")
fig_b.show()
```

```python
plt.plot(data, unit_norm, '.')
plt.ylabel('Unit normed value')
plt.xlabel('Data value')
plt.show()
```

## Scale to new min-max scale

x = x(i) * (new_max - new_min) + new_min


```python
new_min = 20
new_max = 23
restored = unit_norm*(new_max - new_min) + new_min

fig_c, ax_c = plt.subplot_mosaic([['original', 'unity_norm', 're_scaled']], figsize = (12, 4))

ax_c['original'].plot(data, '.')
ax_c['unity_norm'].plot(unit_norm, '.')
ax_c['re_scaled'].plot(restored, '.')

fig_c.show()
```

## Invert the min-max scale
where new scale = a - b

x* = a + ((x - min(x)) / (max(x) - min(x))) * (b - a)

```python
a = 0
b = 10
# To 'invert' the data switch the min and max values to invert the 'direction' of travel
# inverted = a + ((data - min) / (max - min)) * (b - a)
inverted = b + ((data - min) / (max - min)) * (a - b)

fig_d, ax_d = plt.subplot_mosaic([['plot_data', 'plot_inverted']], figsize = (12, 4))

ax_d['plot_data'].plot(data, '.')

ax_d['plot_inverted'].plot(inverted, '.')

fig_d.show()
```
