```python
import numpy as np
import matplotlib.pyplot as plt

from notebooks.helpers import pull
```

```python
n = 40

data = np.random.randn(n, 1)

data[data >= 2] = data[data >= 2]**2
data[data <= -2] = -data[data <= -2]**2
```

## Trim k% or the data

```python
trim_pc = 5
threshold = np.percentile(np.abs(data), 100 - trim_pc)

outliers_pc = np.where(np.abs(data) > threshold)[0]
```

## K most extreme values

```python
k = 3

data_rank = np.argsort(np.abs(data), axis=0)[::-1]
outliers_rank = np.squeeze(data_rank[:k])
```

```python
fig_a, ax_a = plt.subplots()
ax_a.set_xlabel('Data index')
ax_a.set_ylabel('Data value')
ax_a.set_title('Data')

ax_a.plot(outliers_pc, data[outliers_pc], 'rx', markersize=12, label='K% Outliers')
ax_a.plot(outliers_rank, data[outliers_rank], 'ro', markersize=10, alpha=.5, label='Rank Outliers')

ax_a.plot(data, '.')
ax_a.legend()
fig_a.show()
```
