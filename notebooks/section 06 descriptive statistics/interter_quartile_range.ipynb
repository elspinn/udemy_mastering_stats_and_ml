```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
n = 1000
samples = (np.random.randn(n) *10) **2

plt_a, ax_a = plt.subplots(nrows=2,)

ax_a[0].plot(samples, '.')
ax_a[1].hist(samples, bins=30)

plt_a.show()
```

```python
#Ration scale ranked transform
ranked = stats.rankdata(samples) / n

plt_b, ax_b = plt.subplots()
ax_b.plot(samples, ranked, '.')
ax_b.set_ylabel('rank')
ax_b.set_xlabel('sample value')

plt_b.show()
```

```python
#Manual calculation
q1_index = np.argmin((ranked-.25)**2)
q3_index = np.argmin((ranked-.75)**2)

iqr_manual = np.diff([samples[q1_index], samples[q3_index]])

#Scipy method
iqr = stats.iqr(samples)

print(f"IQ1: {samples[q1_index]}, IQ3: {samples[q3_index]}")
print(f"IQR: manual-{iqr_manual[0]}, scipy-{iqr}")
```
