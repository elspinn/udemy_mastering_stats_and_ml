```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

## Gaussian Distribution

```python
N = 1001
x = np.linspace(-4, 4, N)
y = stats.norm.pdf(x)

plt.plot(x, y)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("pdf(x)")

plt.show()
```

```python
# Probability distribution - 0-1

N = 1001
x = np.linspace(-4, 4, N)
y = stats.norm.pdf(x)

x_dist = np.diff(x[:2])
y_area = y * x_dist
total_area = sum(y * x_dist)
y_ration = y_area / total_area


plt.plot(x, y_ration)
plt.title("Normal Distribution")
plt.xlabel("x")
plt.ylabel("pdf(x)")

plt.show()
```

```python
# Normally distributed random numbers

# Parameters
variance = 2
mean = 5
n = 1000

# Create data
data = variance * np.random.randn(n) + mean

# Plot data
plt.hist(data, 25),
plt.title("Empirical Normal Distribution")
plt.show()
```

# Uniform Distribution

```python
# Parameters
variance = 2
mean = 5
n = 1000

# Create Data
data = variance * np.random.rand(n) + mean

# Plot Data
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5, 6))

ax[0].plot(data, '.', markersize=1)
ax[0].set_title("Uniform Data Distribution")

ax[1].hist(data, 25),
ax[1].set_title("Uniform Data Histogram")

fig.show()
```

## Log Normal Distribution

```python
n = 1001
x = np.linspace(0, 10, n)
lognormdist = stats.lognorm.pdf(x, 1)

plt.plot(x, lognormdist)
plt.title("LogNormal Distribution")
plt.show()
```

## Empirical Log Normal Distribution

```python
mean = 5
variance = .5
n = 2000

data = variance * np.random.randn(n) + mean
data = np.exp(data)

print(data)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(5, 6))

ax[0].hist(data, 25)
ax[0].set_title("LogNormal Distribution Histogram")

ax[1].plot(data, '.')
ax[1].set_title("LogNormal Distribution Data")
fig.show()

```

## t

```python
x = np.linspace(-4, 4, 1001)
# Degree of freedom
df = 200
t = stats.t.pdf(x, df)

plt.plot(x, t)
plt.xlabel("t-value")
plt.ylabel('P(t | H$_O$')
plt.title('t(%g) distribution' % df)

plt.show()
```

## F

```python
num_df = 5 # Numerator degrees of freedom
den_df = 100 # Denominator df

x = np.linspace(0, 10, 10001)
df = 40
f_dist = stats.f.pdf(x, num_df, den_df)

plt.plot(x, f_dist)
plt.title(f'F({num_df} {den_df}) Distribution')
plt.xlabel('F value')
plt.show()
```
