```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## Mean

```python
n = 1001
nbins = 30
shift = 2

s1 = 5 * np.random.randn(n)
s2 = np.random.randn(n) + shift
s3 = np.random.randn(n) - shift

fig_a, ax = plt.subplots(nrows=3, ncols=2)

ax[0, 0].plot(s1, '.')
ax[0, 1].hist(s1, bins=nbins)

ax[1, 0].plot(s2, '.')
ax[1, 1].hist(s2, bins=nbins)

ax[2, 0].plot(s3, '.')
ax[2, 1].hist(s3, bins=nbins)

fig_a.show()

fig_b, ax = plt.subplots()
y1, x1 = np.histogram(s1, bins=nbins)
x1 = (x1[1:] + x1[:-1]) / 2

y2, x2 = np.histogram(s2, bins=nbins)
x2 = (x2[1:] + x2[:-1]) / 2

y3, x3 = np.histogram(s3, bins=nbins)
x3 = (x3[1:] + x3[:-1]) / 2

ax.plot(x1, y1, 'b-', label='y1')
ax.plot(x2, y2, 'r-', label='y2')
ax.plot(x3, y3, 'g-', label='y3')

m1 = np.mean(s1)
m2 = np.mean(s2)
m3 = np.mean(s3)

ax.plot([m1, m1], [0, max(y1)], 'b--')
ax.plot([m2, m2], [0, max(y2)], 'r--')
ax.plot([m3, m3], [0, max(y3)], 'g--')

fig_b.show()
```

## Mean for bimodal distribution
Is not an appropriate description of the distribution

```python
shift = 2
n_bins = 30

s4 = np.random.randn(n) + shift
s5 = np.random.randn(n) - shift

s_bimodal = np.hstack((s4, s5))
mean = np.mean(s_bimodal)

y, x = np.histogram(s_bimodal, bins=nbins)
x = (x[1:] + x[:-1]) / 2

fig, ax = plt.subplots()

ax.plot(x, y, 'g-')
ax.plot([mean, mean], [0, max(y)], 'r--')

fig.show()
```

## Median
Median can be a better description of skewed distributions

```python
shift = 0
stretch = .7
n = 2000
n_bins = 40

samples = stretch * np.random.randn(n) + shift
samples = np.exp(samples)

y, x = np.histogram(samples, bins=nbins)
x = (x[1:] + x[:-1]) / 2

median = np.median(samples)
mean = np.mean(samples)

fig_c, ax_c = plt.subplots(nrows=2)

ax_c[0].plot(samples, '.')
ax_c[1].plot(x, y, 'r-')

ax_c[1].plot([median, median], [0, max(y)], 'g--')
ax_c[1].plot([mean, mean], [0, max(y)], 'r--')

fig_c.show()
```

## Mode

```python
data = np.round(np.random.randn(10))

for value in np.unique(data):
    print(f"The value {value} occurs {sum(data == value)} times")

print(f"The mode value is {(stats.mode(data)[0])}")
```
