```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from PIL.GimpGradientFile import linear
```

```python
m = 30
n = 6

normal_dist = np.zeros((m, n))
uniform_dist = np.zeros((m, n))

for i in range(n):
    uniform_dist[:, i] = 30 * np.random.rand(m) * (2 * i / (n - 1) - 1) ** 2 + (i + 1) ** 2
    normal_dist[:, i] = 30 * np.random.randn(m) * (2 * i / (n - 1) - 1) ** 2 + (i + 1) ** 2
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)

ax[0, 0].imshow(normal_dist)
ax[0, 0].set_title('Normal distribution')
ax[0, 1].boxplot(normal_dist)

ax[1, 0].imshow(uniform_dist)
ax[1, 1].boxplot(uniform_dist)
ax[1, 0].set_title('Uniform distribution')
```
