```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import plot_date
```

```python
n = 1000

data1 = np.exp(np.random.randn(n)/2)
data2 = np.exp(np.random.randn(n)/10)
data3 = np.exp(np.random.randn(n)/2 + 1)
```

```python
plt.hist(data1)
plt.hist(data2)
plt.hist(data3)

plt.show()
```

```python
y1, x = np.histogram(data1, bins=40)
x1 = (x[1:] + x[:-1]) /2

y2, x = np.histogram(data2, bins=40)
x2 = (x[1:] + x[:-1]) /2

y3, x = np.histogram(data3, bins=40)
x3 = (x[1:] + x[:-1]) /2

plt.plot(x1, y1, 's-', label='data1')
plt.plot(x2, y2, 'o-', label='data2')
plt.plot(x3, y3, '>-', label='data3')

plt.legend()

plt.show()
```

## Linear vs Logarithmic

```python
x = range(1, 10)
```

```python
fig, axs = plt.subplot_mosaic([['linear', 'logarithmic']])

ax = axs['linear']
ax.plot(x, x, label='x')
ax.plot(x, np.exp(x), label='exp(x)')
ax.legend()

ax= axs['logarithmic']
ax.set_yscale('log')
ax.plot(x, x, label='x')
ax.plot(x, np.exp(x), label='exp(x)')
ax.legend()

fig.show()
```
