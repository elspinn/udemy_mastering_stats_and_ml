```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

## Z Score Transform
Transform data to a standard deviation scale

Algorith
1. Mean center the data
2. Divide by the standard deviation

Formula: z = (x(i) - mean(x)) / std(x)

```python
data = np.random.poisson(lam=1, size=1000)
mean = data.mean()
std = data.std(ddof=1)

plt_a, ax_a = plt.subplot_mosaic([['plot', 'hist']], figsize=[10, 5])
ax_a['plot'].plot(data, '.', alpha=0.5)
ax_a['plot'].set_ylabel('Data value')
ax_a['plot'].set_xlabel('Data index')

ax_a['hist'].hist(data, bins=20)
ax_a['hist'].set_xticks(np.unique(data))
ax_a['hist'].set_ylabel('Count')
ax_a['hist'].set_xlabel('Data value')

plt_a.suptitle(f"Poisson samples\nMean = {np.round(mean, 2)}, Standard Deviation = {np.round(std, 2)}")
plt_a.show()
```

```python
z_scaled = (data - mean)/std
z_mean = z_scaled.mean()
z_std = z_scaled.std(ddof=1)

# z_scaled = stats.zscore(data)
plt_b, ax_b = plt.subplot_mosaic([['plot', 'histogram']], figsize=[10, 5])
ax_b['plot'].plot(z_scaled, '.', alpha=0.5)
ax_b['plot'].set_xlabel('Data index')
ax_b['plot'].set_ylabel('Z score')


ax_b['histogram'].hist(z_scaled, bins=20)
ax_b['histogram'].set_xlabel('Z score')
ax_b['histogram'].set_ylabel('Count')

plt_b.suptitle(f"Z-Indexed\nMean = {np.round(z_mean, 2)}, Standard Deviation = {np.round(z_std, 2)}")
plt_b.show()
```
