```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
```

# Bar Plot

```python
# Create data for the bar plot
m = 30 # rows / samples
n = 6 # columns / attributes

data = np.zeros((m,n))

for i in range(n):
    data[:,i] = 30 * np.random.randn(m) * (2*i/(n-1)-1)**2 + (i+1)**2
```

```python
data
```

```python
fig, ax = plt.subplots(1, 3, figsize=(8, 2))

ax[0].bar(range(n), np.mean(data, axis=0))
ax[0].set_title('Bar plot')

ax[1].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='s', linestyle='')
ax[1].set_title('Error bar')

ax[2].bar(range(n), np.mean(data, axis=0))
ax[2].errorbar(range(n), np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='b')
ax[2].set_title('Error bar plot')

```

```python
xcrossing = [1, 2, 4, 5, 6, 9]
plt.bar(xcrossing, np.mean(data, axis=0))
plt.errorbar(xcrossing, np.mean(data, axis=0), np.std(data, axis=0, ddof=1), marker='.', linestyle='', color='r')
plt.xticks(xcrossing)
plt.title('Bar & error plot')

plt.show()
```

```python
## Note about bars from matrices

m = [[2, 5, 4, 3], [1, 1, 8, 6]]

fix, ax = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

ax[0, 0].imshow(m)

df = pd.DataFrame(m, columns=['a', 'b', 'c', 'd'])
df.plot(ax=ax[1, 0], kind='bar')
ax[1, 0].set_title('Grouping by row')

ax[0, 1].imshow(m)
df.T.plot(ax=ax[1, 1], kind='bar')
ax[1, 1].set_title('Grouping by column')
```
